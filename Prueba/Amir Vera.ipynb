```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
```

2 problemas respecto al ajuste del modelo. Si es por sobre ajuste, el modelo comprende bien los datos para entrenar, pero no llega a predecir otros casos externos, mientras que el subajuste se genera cuando el modelo no logra establecer patrones para predeir datos mas complejo, por lo que sus predicciones son a nivel basico.

En los arboles de decision, los nodos son condiciones a modo de pregunta, las ramas son las respuestas disyuntivas y las hojas se refieren al resultado final. Importante considerar si hay sobreajuste cuando se trabaja con muchas variables.

```python
df = sns.load_dataset('titanic')
df = df[['survived', 'sex', 'pclass', 'age', 'fare']]

# Imputar la edad con la media
df['age'] = df['age'].fillna(df['age'].mean())

# Eliminar solo filas con NA en las otras columnas (fare o sex)
df = df.dropna()

# Codificar 'sex'
df['sex'] = df['sex'].map({'male': 0, 'female': 1})

# Definir X e y
X = df[['sex', 'pclass', 'age', 'fare']]
Y = df['survived']
```

```python
# Dividimos el dataset: 70% entrenamiento, 30% prueba
X_train, X_test, y_train, y_test = train_test_split(
    X, Y, test_size=0.3, random_state=42
)

print("X_train shape:", X_train.shape)
print("X_test shape :", X_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape :", y_test.shape)
```

```python
# 7. Crear y entrenar el árbol de decisión
model = DecisionTreeClassifier(max_depth=3, random_state=42)
model.fit(X_train, y_train)
```

La curva ROC debería tener una tendencia a ser una curva, determina qué tan exitoso resulta el modelo

Una mezcla perfecta ocurre cuando hay una distribucion de 50/50, mientras que el puro ocurre cuando hay un 0 en la variable Gini.
Sample son la cantidad de muestras
Value es el orden de la distribución.
Estos elementos son importantes para comprender la lógica del árbol de decisión

La entropía es el desorden dentro del conjunto de dato, con lo que su indice determina cuan mezcladas estan las clases

Por otra parte, el índice Gini mide la impureza de clases dentro de un nodo, el resultado indica que tan mezcladas estan las clases y la pureza del nodo

Posteriormente se puede realizar una comparación entre ambos para determinar sus niveles de precision

En el pre pruning se delimita la profundidad del arbol mediante condiciones para evitar que el arbol sea muy largo

En el post pruning es el proceso que se realiza luego de que se genere un arbol,con lo que un codigo para obtener el mejor camino posible. Para eso, extrae a los alfa y los pone a prueba para determinar qué tan buenos son.

```python
import pandas as pd
results = []
for depth in [2, 5, None]:
    clf = DecisionTreeClassifier(max_depth=depth, random_state=42)
    clf.fit(X_train, y_train)
    acc_train = clf.score(X_train, y_train)
    acc_test = clf.score(X_test, y_test)
    nodes = clf.tree_.node_count
    results.append({'max_depth': depth or 'None', 'nodes': nodes, 'train_acc': acc_train, 'test_acc': acc_test})
pd.DataFrame(results)
```

En el grafico se busca obtener una interseccion entre la curva de entrenamiento y validación (train ytest). Train empieza desde arriba según la cantidad de veces, mientras que el test comienza con una precisión desde abajo y deben llegar a un punto en común
